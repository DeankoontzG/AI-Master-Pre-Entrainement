```python
#Matrice Spirale :  Ce code est du génie pur, je suis fan. Et je l'ai pas fait moi même malheureusement :/ 


mat = [
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]

def spirale(matrice):
    res = []

    top = 0
    bottom = len(matrice) - 1
    left = 0
    right = len(matrice[0]) - 1

    while top <= bottom and left <= right:
        # ➤ Ligne du haut
        for i in range(left, right + 1):
            res.append(matrice[top][i])
        top += 1

        # ➤ Colonne de droite
        for i in range(top, bottom + 1):
            res.append(matrice[i][right])
        right -= 1

        # ➤ Ligne du bas
        if top <= bottom:
            for i in range(right, left - 1, -1):
                res.append(matrice[bottom][i])
            bottom -= 1

        # ➤ Colonne de gauche
        if left <= right:
            for i in range(bottom, top - 1, -1):
                res.append(matrice[i][left])
            left += 1

    return res

test = spirale(mat)
print(test)
```

```python
#Simulation robot Sherminator simple avec plusieurs structures
#But :  modéliser un robot qui avance sur une grille 2D.
import random
import tkinter as tk
from tkinter import simpledialog

class Robot : 

    def __init__(self, nom) :
        self.nom = nom
        self.historikPos = [(10,10)]
        self.visitedPos = set({(10,10)})
        self.obstacles = set()
        self.worldsize = 20  

    def ajoutObstacles(self, *args) :
        if len(args) % 2 != 0:
            print("⚠️ Nombre impair d'entiers pour obstacles : le dernier sera ignoré.")
            args = args[:-1]  # supprime le dernier

        for i in range(0, len(args), 2):
            self.obstacles.add((args[i], args[i+1]))

    def printObstacles(self) : 
        print(self.obstacles)
        
    def majPos(self, x, y) :
        self.historikPos.append((x, y))
        if (x, y) in self.visitedPos : 
            print("Déjà visité mon pote !") 
        self.visitedPos.add((x, y))

    def choixdir(self):
        dir = random.randint(1, 4)
        x, y = self.historikPos[-1]
        if dir == 1:
            y += 1  # Haut
        elif dir == 2:
            x += 1  # Droite
        elif dir == 3:
            y -= 1  # Bas
        elif dir == 4:
            x -= 1  # Gauche
        else:
            print("Erreur dans choixdir : valeur inattendue")

        return(self.actionAvancer(x, y))
    
    def actionAvancer(self, x, y):
        if (x, y) in self.obstacles:
            return(f"{x} / {y} est un obstacle, avancée bloquée")
        elif x < 0 or y < 0 or x >= self.worldsize or y >= self.worldsize:
            return("Limites du monde atteintes, pas d'action ce tour-ci")
        else:
            self.majPos(x, y)
            return(f"Avancée vers {x} / {y}")
            


# --- Interface Tkinter ---

class InterfaceRobot(tk.Tk):
    def __init__(self, robot):
        super().__init__()
        self.title(f"Grille de {robot.nom}")
        self.robot = robot
        self.cell_size = 20
        self.canvas = tk.Canvas(self, width=robot.worldsize * self.cell_size,
                                     height=robot.worldsize * self.cell_size)
        self.canvas.pack()

        # Zone de texte pour logs
        self.text_output = tk.Text(self, height=4, state='disabled', bg="lightyellow")
        self.text_output.pack(fill="x")

        # Boutons
        bouton_frame = tk.Frame(self)
        bouton_frame.pack(pady=5)

        tk.Button(bouton_frame, text="➕ Ajouter obstacle", command=self.ajouter_obstacle).pack(side="left", padx=10)
        tk.Button(bouton_frame, text="🚶 Avancer robot", command=self.deplacer_robot).pack(side="left", padx=10)

        self.dessiner_grille()

    def log(self, message):
        self.text_output.config(state='normal')
        self.text_output.insert(tk.END, message + '\n')
        self.text_output.see(tk.END)
        self.text_output.config(state='disabled')

    def dessiner_grille(self):
        self.canvas.delete("all")
        taille = self.robot.worldsize

        for y in range(taille):
            for x in range(taille):
                color = "white"
                if (x, y) in self.robot.obstacles:
                    color = "black"
                elif (x, y) in self.robot.visitedPos:
                    color = "lightgrey"
                if (x, y) == self.robot.historikPos[-1]:
                    color = "red"

                self.canvas.create_rectangle(
                    x * self.cell_size, (taille - y - 1) * self.cell_size,
                    (x + 1) * self.cell_size, (taille - y) * self.cell_size,
                    fill=color, outline="gray"
                )

    def ajouter_obstacle(self):
        # Boîte de saisie utilisateur
        input_str = simpledialog.askstring(
            "Ajouter obstacles",
            "Entrez des coordonnées x,y sous forme d'entiers séparés par des virgules (ex: 1,4,5,6):"
        )
        
        if input_str is None:
            return  # L'utilisateur a annulé

        try:
            # Transformation en liste d'entiers
            valeurs = [int(val.strip()) for val in input_str.split(",") if val.strip().isdigit()]

            if len(valeurs) < 2:
                self.log("⚠️ Pas assez de coordonnées.")
                return

            # Appel de la méthode du robot
            self.robot.ajoutObstacles(*valeurs)
            self.log(f"Obstacles ajoutés avec : {valeurs}")
            self.dessiner_grille()

        except ValueError:
            self.log("❌ Erreur : veuillez entrer uniquement des entiers valides séparés par des virgules.")

    def deplacer_robot(self):
        msg = self.robot.choixdir()
        self.log(msg)
        self.dessiner_grille()

Sherminator = Robot("Sherminator")
Sherminator.ajoutObstacles(1, 8, 7, 6, 5, 4, 3, 2, 1, 4, 6, 5)
app = InterfaceRobot(Sherminator)
app.mainloop()
```
